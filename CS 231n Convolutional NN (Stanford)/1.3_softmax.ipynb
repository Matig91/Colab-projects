```python
def get_CIFAR10_data(num_training=49000, num_validation=1000, num_test=1000, num_dev=500):
    """    Load the CIFAR-10 dataset from disk and perform preprocessing to prepare
    it for the linear classifier. These are the same steps as we used for the
    SVM, but condensed to a single function. """

    # Load the raw CIFAR-10 data
    cifar10_dir = 'cs231n/datasets/cifar-10-batches-py'

    # Cleaning up variables to prevent loading data multiple times (which may cause memory issue)
    try:
       del X_train, y_train
       del X_test, y_test
       print('Clear previously loaded data.')
    except:
       pass

    X_train, y_train, X_test, y_test = load_CIFAR10(cifar10_dir)

    # subsample the data
    mask = list(range(num_training, num_training + num_validation))
    X_val = X_train[mask]
    y_val = y_train[mask]
    mask = list(range(num_training))
    X_train = X_train[mask]
    y_train = y_train[mask]
    mask = list(range(num_test))
    X_test = X_test[mask]
    y_test = y_test[mask]
    mask = np.random.choice(num_training, num_dev, replace=False)
    X_dev = X_train[mask]
    y_dev = y_train[mask]

    # Preprocessing: reshape the image data into rows
    X_train = np.reshape(X_train, (X_train.shape[0], -1))
    X_val = np.reshape(X_val, (X_val.shape[0], -1))
    X_test = np.reshape(X_test, (X_test.shape[0], -1))
    X_dev = np.reshape(X_dev, (X_dev.shape[0], -1))

    # Normalize the data: subtract the mean image
    mean_image = np.mean(X_train, axis = 0)
    X_train -= mean_image
    X_val -= mean_image
    X_test -= mean_image
    X_dev -= mean_image

    # add bias dimension and transform into columns
    X_train = np.hstack([X_train, np.ones((X_train.shape[0], 1))])
    X_val = np.hstack([X_val, np.ones((X_val.shape[0], 1))])
    X_test = np.hstack([X_test, np.ones((X_test.shape[0], 1))])
    X_dev = np.hstack([X_dev, np.ones((X_dev.shape[0], 1))])

    return X_train, y_train, X_val, y_val, X_test, y_test, X_dev, y_dev


# Invoke the above function to get our data.
X_train, y_train, X_val, y_val, X_test, y_test, X_dev, y_dev = get_CIFAR10_data()
print('Train data shape: ', X_train.shape)
print('Train labels shape: ', y_train.shape)
print('Validation data shape: ', X_val.shape)
print('Validation labels shape: ', y_val.shape)
print('Test data shape: ', X_test.shape)
print('Test labels shape: ', y_test.shape)
print('dev data shape: ', X_dev.shape)
print('dev labels shape: ', y_dev.shape)
```

## Softmax Classifier

```python
from builtins import range
import numpy as np
from random import shuffle
from past.builtins import xrange

def softmax_loss_naive(W, X, y, reg):
    """   Softmax loss function, naive implementation (with loops)
    Inputs have dimension D, there are C classes, and we operate on minibatches of N examples.
    Inputs:  - W: A numpy array of shape (D, C) containing weights.
    - X: A numpy array of shape (N, D) containing a minibatch of data.
    - y: A numpy array of shape (N,) containing aining labels; y[i] = c means
      that X[i] has label c, where 0 <= c < C.  - reg: (float) regularization strength
    Returns a tuple of:  - loss as single float
    - gradient with respect to weights W; an array of same shape as W           """
    # Initialize the loss and gradient to zero
    loss = 0.0; dW = np.zeros_like(W)
    """ # TODO: Compute the softmax loss and its gradient using explicit loops
    # Store the loss in loss and the gradient in dW. If you are not careful     #
    # here, it is easy to run into numeric instability. Don't forget the        #
    # regularization!       """

    scores = X.dot(W)
    num_classes = W.shape[1]
    num_train   = X.shape[0]
    for i in range(num_train):
      f = scores[i] - np.max(scores[i]) # num stability
      softmax = np.exp(f)/np.sum(np.exp(f))
      # print(f.shape)
      loss     += -np.log(softmax[y[i]])
      for j in range(num_classes):

        dW[:,j]  += X[i] * softmax[j]
      dW[:,y[i]] -= X[i]

    # Average
    loss /= num_train;  dW   /= num_train
    # Reg-n
    loss += reg * np.sum(W*W)
    dW   += 2 * reg * W

    return loss, dW

def softmax_loss_vectorized(W, X, y, reg):
    """ Softmax loss function, vectorized version.
    Inputs and outputs are the same as softmax_loss_naive. """
    # Initialize the loss and gradient to zero

    loss = 0.0
    dW = np.zeros_like(W)
    """  # TODO: Compute the softmax loss and its gradient using no explicit loops.
    Store the loss in loss and the gradient in dW. If you are not careful
    here, it is easy to run into numeric instability. Don't forget the regularization!  """

    scores      = X.dot(W)
    num_classes = W.shape[1]
    num_train   = X.shape[0]

    scores = scores - np.max(scores, axis=1, keepdims=True)                          # num stability
    softmax = np.exp(scores)/np.sum(np.exp(scores), axis=1, keepdims=True)

    loss = -np.sum(np.log(softmax[np.arange(num_train),y]), keepdims=True)

    dW           =  np.matmul(X.T, softmax)             # dim x class = X.T(dim x num) | num x class
    yenc         = np.zeros((num_train, num_classes))
    yenc[np.arange(num_train), y] = 1
    dW          -= X.T @ yenc   # dim x class = dim x num | num x class = X.T @ (num x 1)
    # Average
    loss /= num_train;  dW   /= num_train

    # Reg-n
    loss += reg * np.sum(W*W)
    dW   += 2 * reg * W

    return loss, dW

def softmax_loss_vectorized_2(W, X, y, reg):
    loss = 0.0;    dW = np.zeros_like(W)

    scores = X.dot(W)

    num_classes = W.shape[1]
    num_train   = X.shape[0]
    scores = scores - np.max(scores, axis=1, keepdims=True)                          # num stability

    softmax = np.exp(scores)/np.sum(np.exp(scores), axis=1, keepdims=True)
    loss = -np.sum(np.log(softmax[np.arange(num_train),y]), keepdims=True)
    softmax[np.arange(num_train), y] += -1              # Softmax Matrix
    dW           =  np.matmul(X.T, softmax)             # dim x class = X.T(dim x num) | num x class

    loss /= num_train;  dW   /= num_train
    loss += reg * np.sum(W*W);    dW   += 2 * reg * W
    return loss, dW
```
